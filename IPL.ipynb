```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl=pd.read_csv("D:\Download\matches.csv")
```

```python
ipl.tail()
```

```python
ipl.shape
```

# Most MOTMs

```python
ipl['player_of_match'].value_counts()
```

```python
ipl['player_of_match'].value_counts().head(1)
```

```python
sns.catplot(y='player_of_match',data=ipl,kind='count',order=ipl['player_of_match'].value_counts().head(10).index,color='r')
plt.title('MOTMS_DISTRIBUTION')
plt.show
```

# Win by Teams

```python
plt.figure(figsize=(50, 20))
ipl['winner'].value_counts().plot.pie(autopct='%1.1f%%',startangle=90, 
 colors=['blue', 'yellow', 'red', 'purple'],wedgeprops={'edgecolor': 'black','linewidth':1.5})
```

# Most Toss Win

```python
plt.figure(figsize=(50, 20))
sns.histplot(ipl['toss_winner'],bins=10,color='purple')
```

```python
ipl['city'].isnull().sum()
```

```python
ipl['city'].value_counts()
```

```python
# win by runs
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
# making ahistogram
plt.figure(figsize=(5,7))
plt.hist(batting_first['win_by_runs'])
plt.title('Runs_Distribution')
plt.xlabel("Runs")
plt.show()
```

```python
batting_first['winner'].value_counts()
```

```python
# making a barchat
plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts().head(3).keys()),list(batting_first['winner'].value_counts().head(3)),color=["blue","green","Red"])
plt.show()
```

```python
# making a pichar
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
bowling_first=ipl[ipl["win_by_wickets"]!=0]
bowling_first.head()
```

```python
plt.hist(bowling_first['win_by_wickets'],bins=30)
plt.title('wicket_Distrubation')
plt.xlabel("wicket")
plt.show()
```

```python
#hbar on win by 2nd batting
plt.figure(figsize=(10,10))
plt.bar(list(bowling_first['winner'].value_counts().head(5).keys()),list(bowling_first['winner'].value_counts().head()),color=['blue','yellow','black','cyan','red'])
plt.show()
```

```python
#piechart on 2nd batting
plt.figure(figsize=(10,10))
plt.pie(list(bowling_first['winner'].value_counts()),labels=list(bowling_first['winner'].value_counts().keys()),autopct='%1.1f%%')
```
